```python
invalid = "Invalid input. Please enter 'y' for yes or 'n' for no."
invalidNumber = "The input was invalid, please input a valid number."
enter = "Enter 'y' for yes or 'n' for no."
yesCompare = ["y", "ye", "yes"]
noCompare = ["n", "no"]


#Step 1: Find your monthly/annual income
#Step 2: Find you monthly/annual expenses
#Step 3: Calculate your cash flow
#Step 4: Calculate your ROI


#Step 1: Start with Income
#Input for user services
laundryPayment = input("Does the tenant pay for laundry services? " + enter)
if laundryPayment == "n":
    laundryPayment = 0
elif laundryPayment == "y":
    laundryPayment = int(input("What is the tenant paying for laundry services?"))
else:
    print(invalid)

storagePayment = input("Does the tenant pay for storage services? " + enter)
if storagePayment == "n":
    storagePayment = 0
elif storagePayment == "y":
    storagePayment = int(input("What is the tenant paying for storage services?"))
else:
    print(invalid)

tenantMiscExpenses = input("Are there any other services the tenant is paying for? " + enter)
if tenantMiscExpenses == "n":
    tenantMiscExpenses = 0
elif tenantMiscExpenses == "y":
    tenantMiscExpenses = int(input("What is the tenant paying for these services?"))
else:
    print(invalid)

#User input for tenant rent
tenantRent = int(input("What is the amount of rent the current tenant is paying?"))

#Calculation for tenant services(laundry, storage, misc)
tenantServices = laundryPayment + storagePayment + tenantMiscExpenses

#Income is calculated through combination of tenant paid rent and any services the tenant is using
siteIncome = tenantRent + tenantServices
siteIncomeAnnual = siteIncome * 12

print(f"Your site's current monthly income is ", "${:,.2f}".format(siteIncome))
print(f"Your site's current annual income is ", "${:,.2f}".format(siteIncomeAnnual))

#Step 2: Calculate expenses
#Tax & Insurance (escrow), Utilities (Electric, Water, Sewer, Garbage, Gas), HOA, Lawn/Snow, Vacancy, Repairs, Capital Expenses, Property Manager, Mortgage

#Escrow
escrow = input("Do you have an Escrow account to cover property taxes and mortgage insurance or do you pay property taxes and mortgage insurance seperately? Press '1' for Escrow account or '2' for separate funds")
if escrow == "1":
    escrow = int(input("How much is the monthly Escrow cost?"))
    escrowAnnual = escrow * 12
    print(f"Your monthly Escrow cost is ", "${:,.2f}".format(escrow))
    print(f"Your annual Escrow cost is ", "${:,.2f}".format(escrowAnnual))
elif escrow == "2":
    propertyTax = int(input("What are your expected annual property taxes for this site?"))/12
    siteInsurance = input("Are you setting aside money for property insurance for this location? " + enter)
    if siteInsurance == "y":
        siteInsurance = int(input("How much is set aside each month for property insurance?"))
    elif siteInsurance == "n":
        siteInsurance = 0
    else: 
        print(invalid)        
    escrow = propertyTax + siteInsurance
    escrowAnnual = escrow * 12
    print(f"Your expected monthly savings for property taxes and mortgage insurance is ", "${:,.2f}".format(escrow))
    print(f"Your expected annual savings for for property taxes and mortgage insurance is ", "${:,.2f}".format(escrowAnnual))
else:
    print(invalid)

#Utilies
#electric
electric = input("Do you have a monthly electric bill? " + enter)
if any(electric.lower() == f for f in yesCompare):
    electric = int(input("What is your average monthly cost for electricity?"))
    print(f"Your estimated monthly electric bill is", "${:,.2f}".format(electric))
elif any(electric.lower() == f for f in noCompare):
    electric = 0
    print("You elected that you do not have an electric bill.")
else:
    print(invalid)

#water
water = input("Do you have a monthly recurring water bill? " + enter)
if any(water.lower() == f for f in yesCompare):
    water = int(input("What is your average monthly cost for water?"))
    print(f"Your water bill is", "${:,.2f}".format(water))
elif any(water.lower() == f for f in noCompare):
    water = 0
    print("You elected that you do not have a water bill.")
else:
    print(invalid)


#Gas
gas = input("Do you have a monthly recurring gas bill? " + enter)
if any(gas.lower() == f for f in yesCompare):
    gas = int(input("What is your average monthly cost for this bill?"))
    print(f"Your gas bill is", "${:,.2f}".format(gas))
elif any(gas.lower() == f for f in noCompare):
    gas = 0
    print("You elected that you do not have a gas bill.")
else:
    print(invalid)

#Sewage
sewage = input("Do you have a monthly recurring sewage bill? " + enter)
if any(sewage.lower() == f for f in yesCompare):
    sewage = int(input("What is your average monthly cost for this bill?"))
    print(f"Your sewage bill is", "${:,.2f}".format(sewage))
elif any(sewage.lower() == f for f in noCompare):
    sewage = 0
    print("You elected that you do not have a sewage bill.")
else:
    print(invalid)

#Garbage
garbage = input("Do you have a monthly recurring trash collection bill? " + enter)
if any(garbage.lower() == f for f in yesCompare):
    garbage = int(input("What is your average monthly cost for this bill?"))
    print(f"Your garbage bill is", "${:,.2f}".format(garbage))
elif any(garbage.lower() == f for f in noCompare):
    garbage = 0
    print("You elected that you do not have a trash collection bill.")
else:
    print(invalid)

utilityBill = (electric + water + gas + sewage + garbage)
utilityBillAnnual = utilityBill * 12

print(f"Your total monthly utilities bill is " + "${:,.2f}".format(utilityBill))
print(f"Your total annual utilities bill is " + "${:,.2f}".format(utilityBillAnnual))

#HOA
hoa = input("Is there an HOA present? " + enter)
if any(hoa.lower() == f for f in yesCompare):
    hoa = input("Are there any monthly fees for this HOA? " + enter)
    if any(hoa.lower() == f for f in yesCompare):
        hoa = int(input("What is your average monthly cost for this bill?"))
        hoaAnnual = hoa * 12
        print(f"Your monthly HOA membership expense is", "${:,.2f}".format(hoa))
        print(f"Your annual HOA membership expense is", "${:,.2f}".format(hoaAnnual))
    elif any(hoa.lower() == f for f in noCompare):
        hoa = 0
        hoaAnnual = 0
        print("How did you swing that?")
    else:
        print(invalid)
elif any(hoa.lower() == f for f in noCompare):
    hoa = 0
    hoaAnnual = 0
    print("Congradulations! Good on you finding a place without those bloodsuckers!")
else:
    print(invalid)

#Lawn/Snow
lawnAndSnow = input("Are there any charges related to monthly lawn care or snow removal? " + enter)
if any(lawnAndSnow.lower() == f for f in yesCompare):
    lawnFrequency = input("Are there expected charges related to lawn care? " + enter)
    if any(lawnFrequency.lower() == f for f in yesCompare):
        lawnFrequency = int(input("How many occurances of lawn care are expected for the year?"))
        lawnCharge = int(input("How much is the cost of the current lawn care per occurance?"))
    elif any(lawnFrequency.lower() == f for f in noCompare):
        lawnFrequency = 0
        lawnCharge = 0
    else:
        lawnFrequency = 0
        lawnCharge = 0
        print(invalid)
    lawnCalc = lawnFrequency * lawnCharge
    #snow
    snowFrequency = input("Are there expected charges related to snow removal? " + enter)
    if any(snowFrequency.lower() == f for f in yesCompare):
        snowFrequency = int(input("How many occurances of snow removal are expected for the year?"))
        snowCharge = int(input("How much is the cost of snow removal per occurance?"))
    elif any(snowFrequency.lower() == f for f in noCompare):
        snowFrequency = 0
        snowCharge = 0
    else:
        print(invalid)
        snowCalc = 0
    snowCalc = snowFrequency * snowCharge
    lawnAndSnow = (lawnCalc + snowCalc)
    lawnAndSnowMonthly = lawnAndSnow / 12
    print(f"Your expected monthly lawn care and snow removal costs are " + "${:,.2f}".format(lawnAndSnowMonthly))
    print(f"Your expected annual lawn care and snow removal costs are " + "${:,.2f}".format(lawnAndSnow))
elif any(lawnAndSnow.lower() == f for f in noCompare):
    lawnAndSnow = 0
    print("You have no expected lawn and snow removal costs for the year")
else:
    print(invalid)


#vacancy
vacancy = int(input("How many months of vacancy do you want to have prepared? (Minimum 2)"))
if vacancy >= 2:
    vacancy = vacancy * siteIncome
else:
    print("You have selected an invalid response, please pick a number with a value of 2 or greater.")

print(f"Your recommended total vacancy coverage is ", "${:,.2f}".format(vacancy))

#Repairs
homeValue = int(input("What is the current value of your home? (Enter a number with no commas or symbols)"))
repairs = int(input("Recommended savings is between 1-4%. Please enter a number 1-4 to designate your preferred savings amount"))
if repairs < 1:
    print(invalidNumber)
elif repairs > 4:
    print(invalidNumber)
else:
    repairs = homeValue * (repairs / 100)

print("Your recommended repair savings amount for the year is ", "${:,.2f}".format(repairs))


#Capital Expenses
capitalExpenses = input("Would you like to set aside an amount each month for capital expenses? " + enter)
if any(capitalExpenses.lower() == f for f in yesCompare):
    capitalExpenses = int(input("How much would you like to set aside each month for capital expenses?"))
    if capitalExpenses >= 0:
        capitalExpenses = capitalExpenses
    else: 
        capitalExpenses = 0
        print(invalidNumber)
    capitalExpensesAnnual = capitalExpenses * 12
    print(f"Your annual savings for capital expenses should be ", "${:,.2f}".format(capitalExpensesAnnual))
elif any(capitalExpenses.lower() == f for f in noCompare):
    capitalExpenses = 0
    capitalExpensesAnnual = capitalExpenses * 12
    print(f"You have elected to not save for capital expenses")
else: 
    print(invalid)

#Property Manager
propertyManager = input("Do you have a property manager for this site? " + enter)
if any(propertyManager.lower() == f for f in yesCompare):
    propertyManager = int(input("How much do you currently pay your property manager on a monthly basis?"))
    propertyManagerAnnual = propertyManager * 12
    print(f"Your annual property management cost is ", "${:,.2f}".format(propertyManagerAnnual))    
elif any(propertyManager.lower() == f for f in noCompare):
    propertyManager = 0
    propertyManagerAnnual = propertyManager * 12
    print("You do not have any property Manager expenses")
else: 
    print(invalid)

#Mortgage
mortgage = int(input("What are your current monthly mortgage payment?"))
mortgageAnnual = mortgage * 12
print(f"Your annual mortgage payment is ", "${:,.2f}".format(mortgageAnnual))

siteExpensesAnnual = escrowAnnual + utilityBillAnnual + hoaAnnual + lawnAndSnow + vacancy + repairs + capitalExpensesAnnual + propertyManagerAnnual + mortgageAnnual
siteExpenses = siteExpensesAnnual / 12

print(f"Your monthly site expenses are ", "${:,.2f}".format(siteExpenses))
print(f"Your annual site expenses are ", "${:,.2f}".format(siteExpensesAnnual))


#Step 3: Find that cash flow
cashFlow = siteIncome - siteExpenses
cashFlowAnnual = siteIncomeAnnual - siteExpensesAnnual

print(f"Your monthly cash flow is ", "${:,.2f}".format(cashFlow))
print(f"Your annual cash flow is ", "${:,.2f}".format(cashFlowAnnual))

#Calculate ROI
downPayment = input("Was there a down payment on the site? " + enter)
if any(downPayment.lower() == f for f in yesCompare):
    downPayment = int(input("How much was the down payment for the site?"))
    print(f"The down payment for the house was ", "${:,.2f}".format(downPayment))
elif any(downPayment.lower() == f for f in noCompare):
    downPayment = 0
    print("There was not a down payment on this site.")
else: 
    print(invalid)

closingCosts = input("Were there closing costs associated with the site purchase? " + enter)
if any(closingCosts.lower() == f for f in yesCompare):
    closingCosts = int(input("How much were the closing costs for the site purchase?"))
    print(f"The closing costs associated with this site were ", "${:,.2f}".format(closingCosts))
elif any(closingCosts.lower() == f for f in noCompare):
    closingCosts = 0
    print("There were no closing costs for this site.")
else: 
    print(invalid)

rehab = input("Were there any costs associated with refurbishing or rehabilitating the site? " + enter)
if any(rehab.lower() == f for f in yesCompare):
    rehab = int(input("How much was used for site rehabilitation?"))
    print(f"The costs associated with site rehabilitation were ", "${:,.2f}".format(rehab))
elif any(rehab.lower() == f for f in noCompare):
    rehab = 0
    print("There were no rehabilitation costs associated with this site.")
else: 
    print(invalid)

siteMisc = input("Were there any other costs associated with or invested into this site? " + enter)
if any(siteMisc.lower() == f for f in yesCompare):
    siteMisc = int(input("How much was invested into this site?"))
    print(f"The costs associated with site investments were ", "${:,.2f}".format(siteMisc))
elif any(siteMisc.lower() == f for f in noCompare):
    siteMisc = 0
    print("There were no extra investments associated with this site.")
else: 
    print(invalid)

totalInvestment = downPayment + closingCosts + rehab + siteMisc
print(f"Your total site investment is ", "${:,.2f}".format(totalInvestment))

roi = (cashFlowAnnual / totalInvestment )
print(f"Your cash on cash ROI is ", "{:.2%}".format(roi))
```
